```python
sns.barplot(x='sex', y='total_bill', data=tips_data) #totalbills are mostly paid by the males
```

```python
sns.distplot(tips_data['total_bill'])
```

```python
sns.jointplot(x = 'total_bill', y='tip', data=tips_data, kind='hex') #darker regions means more number of points are present
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='reg') #the tip value increases based on the total bill
```

```python
sns.pairplot(tips_data)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_data) #most bills are paid on saturday and sunday
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_data, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill',data=tips_data) #draws scatter points of a variable 
```
